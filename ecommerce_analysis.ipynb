```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

print("Libraries imported successfully")
```

```python
df = pd.read_excel("../dataset/Online Retail.xlsx")
df.head()
```

```python
df.shape
```

```python
df.info()
```

```python
df.columns
df.info()
```

```python
df.head(10)
```

```python
df.isnull().sum()
```

```python
df.describe()
```

```python
# removing unwanted data
df = df.dropna(subset=['CustomerID', 'Description'])
```

```python
df.isnull().sum()
```

```python
# check and remove duplicates
df.duplicated().sum()
```

```python
df = df.drop_duplicates()
```

```python
df.duplicated().sum()
```

```python
#remove canceled records
df[df['InvoiceNo'].astype(str).str.startswith('C')].shape
```

```python
df = df[~df['InvoiceNo'].astype(str).str.startswith('C')]
```

```python
df[df['InvoiceNo'].astype(str).str.startswith('C')].shape
```

```python
# remove negative data
(df['Quantity'] < 0).sum()
```

```python
(df['UnitPrice'] < 0).sum()
```

```python
df = df[(df['Quantity'] > 0) & (df['UnitPrice'] > 0)]
```

```python
(df['Quantity'] < 0).sum()
(df['UnitPrice'] < 0).sum()
```

```python
(df['UnitPrice'] < 0).sum()
```

```python
# total price column (new)
df['TotalPrice'] = df['Quantity'] * df['UnitPrice']
```

```python
df.head()
```

```python
# Convert InvoiceDate to Datetime Format
df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'])
```

```python
# month(new)
df['Month'] = df['InvoiceDate'].dt.month
df['Day'] = df['InvoiceDate'].dt.day
df['Hour'] = df['InvoiceDate'].dt.hour
```

```python
df.head()
```

```python
monthly_sales = df.groupby('Month')['TotalPrice'].sum()

monthly_sales
```

```python
monthly_sales.plot(kind='bar', figsize=(10,5))

plt.title('Monthly Sales Revenue')
plt.xlabel('Month')
plt.ylabel('Revenue')

plt.show()
```

```python
top_products = df.groupby('Description')['Quantity'].sum()
top_products = top_products.sort_values(ascending=False)
top_products.head(10)
```

```python
top_products.head(10).plot(
    kind='bar',
    figsize=(14,6),
    color='skyblue'
)

plt.title('Top 10 Selling Products', fontsize=16)
plt.xlabel('Products', fontsize=12)
plt.ylabel('Quantity Sold', fontsize=12)

plt.xticks(rotation=75)

plt.tight_layout()

plt.show()
```

```python
country_sales = df.groupby('Country')['TotalPrice'].sum()
country_sales = country_sales.sort_values(ascending=False)
country_sales.head(10)
```

```python
country_sales.head(10).plot(
    kind='bar',
    figsize=(12,6),
    color='orange'
)

plt.title('Top 10 Revenue Generating Countries', fontsize=16)
plt.xlabel('Country', fontsize=12)
plt.ylabel('Revenue', fontsize=12)

plt.xticks(rotation=45)

plt.tight_layout()

plt.show()
```

```python
hourly_sales = df.groupby('Hour')['TotalPrice'].sum()
hourly_sales.plot(
    kind='line',
    figsize=(12,6),
    marker='o',
    color='green'
)

plt.title('Hourly Sales Revenue', fontsize=16)
plt.xlabel('Hour of Day', fontsize=12)
plt.ylabel('Revenue', fontsize=12)

plt.grid(True)

plt.show()
```

```python
top_customers = df.groupby('CustomerID')['TotalPrice'].sum()
top_customers = top_customers.sort_values(ascending=False)
top_customers.head(10)
```

```python
top_customers.head(10).plot(
    kind='bar',
    figsize=(12,6),
    color='purple'
)

plt.title('Top 10 Customers by Revenue', fontsize=16)
plt.xlabel('Customer ID', fontsize=12)
plt.ylabel('Revenue', fontsize=12)

plt.xticks(rotation=45)

plt.tight_layout()

plt.show()
```

```python
correlation = df[['Quantity', 'UnitPrice', 'TotalPrice']].corr()

correlation
```

```python
plt.figure(figsize=(8,6))

sns.heatmap(
    correlation,
    annot=True,
    cmap='coolwarm'
)

plt.title('Correlation Heatmap')

plt.show()
```
